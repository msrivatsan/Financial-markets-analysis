```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
path1 = '/content/drive/MyDrive/Dataset/coin_Bitcoin.csv'
path2 = '/content/drive/MyDrive/Dataset/coin_Ethereum.csv'
```

```python
df1 = pd.read_csv(path1)
df2 = pd.read_csv(path2)
df1.shape
```

```python
df2.shape
```

```python
print(df1.head())
print(df2.head())
```

```python
print(df1.info())
print(df2.info())
```

```python
# converting date column to datetime format
df1['Date'] = pd.to_datetime(df1['Date'])
df2['Date'] = pd.to_datetime(df2['Date'])
df1.head(5)
```

```python
# sorting the data by date
df1.sort_values('Date', inplace=True)
df2.sort_values('Date', inplace=True)
df1.head(5)
```

```python
# checking for null values
print(df1.isnull().sum())
print(df2.isnull().sum())
```

```python
# Checking data types
print(df1.dtypes)
print(df2.dtypes)
```

```python
# Dropping null values and making that true for the entire dataset
df1.dropna(inplace=True)
df2.dropna(inplace=True)
```

```python
print(df1.head(5))
print(df2.head(5))
print(df1.shape)
print(df2.shape)
```

```python
# Creating a daily change column, which shows the % difference in price from the previous day
df1['Daily Change'] = df1['Close'].pct_change()
df2['Daily Change'] = df2['Close'].pct_change()
print(df1.head(5))
```

```python
df1['Daily Change'] = df1['Daily Change'] * 100
df2['Daily Change'] = df2['Daily Change'] * 100
```

```python
# to get rid of NaN value in first row of daily change column
df1.dropna(subset=['Daily Change'], inplace=True)
df2.dropna(subset=['Daily Change'], inplace=True)
print(df1.head(2))
print(df2.head(2))
```

```python
# First piece of analysis, showing the frequency of different percentages of daily changes
histx = df1['Daily Change']
plt.hist(histx, bins=100, color='blue', alpha = 0.7)
plt.xlabel('Daily Change (%)')
plt.ylabel('Frequency')
plt.title('Bitcoin Daily Changes')

plt.grid(color = 'red', linestyle = '-.', axis = 'y')

plt.show()

histx2 = df2['Daily Change']
plt.hist(histx2, bins=100, color='blue', alpha = 0.7)
plt.xlabel('Daily Change (%)')
plt.ylabel('Frequency')
plt.title('Ethereum Daily Changes')

plt.grid(color = 'red', linestyle = '-.', axis = 'y')

plt.show()
```

```python
# Box plot of daily changes to show which Cryptocurrency has higher volatility
boxx = df1['Daily Change']
boxy = df2['Daily Change']

plt.boxplot([boxx, boxy])
plt.xticks([1, 2], ['Bitcoin', 'Ethereum'])
plt.xlabel('Cryptocurrency')
plt.ylabel('Daily Change (%)')
plt.title('Daily Change Box Plot')

plt.grid(color = 'grey', linestyle = "--")

plt.show()
```

```python
# Continuing data analysis - analysis of average prices over 30 and 100 days
df1['A_30d'] = df1['Close'].rolling(window=30).mean()
df1['A_100d'] = df1['Close'].rolling(window=100).mean()
```

```python
# Plotting closing price vs 30 day average vs 100 day average for bitcoin
xpoints = df1['Date']
ypoints = df1['Close']
plt.plot(xpoints, ypoints, label = 'Closing Price')

xpoints1 = df1['Date']
ypoints1 = df1['A_30d']
plt.plot(xpoints1, ypoints1, label = '30 day average')

xpoints2 = df1['Date']
ypoints2 = df1['A_100d']
plt.plot(xpoints2, ypoints2, label = '100 day average')

plt.legend()
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.title('Bitcoin Closing Price and Averages')

plt.grid(color = 'grey', linestyle = "--")

plt.show()
```

```python
df2['A_30d'] = df2['Close'].rolling(window=30).mean()
df2['A_100d'] = df2['Close'].rolling(window=100).mean()
```

```python
# Plotting closing price vs 30 day average vs 100 day average for ethereum
xpoints = df2['Date']
ypoints = df2['Close']
plt.plot(xpoints, ypoints, label = 'Closing Price')

xpoints1 = df2['Date']
ypoints1 = df2['A_30d']
plt.plot(xpoints1, ypoints1, label = '30 day average')

xpoints2 = df2['Date']
ypoints2 = df2['A_100d']
plt.plot(xpoints2, ypoints2, label = '100 day average')

plt.legend()
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.title('Ethereum Closing Price and Averages')

plt.grid(color = 'grey', linestyle = "--")

plt.show()
```
